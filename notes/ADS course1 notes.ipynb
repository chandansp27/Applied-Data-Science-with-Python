In python the map() function returns a map object it doesnt run the function in the map arguement. This helps in efficient memory management.

```python
#map(function, iterable)
store1 = [10.00, 11.00, 12.98, 2.34]
store2 = [9.00, 11.10, 12.04, 2.01]
cheapest = map(min, store1, store2)

# square an element using a list
cheapest_list = list(cheapest)
squares_cheapset = map(lambda x: x**2, cheapest_list)

[f(x) for f in [lambda x:'%.2f'%x] for x in  squares_cheapset]
```

##### Quantifiers

Quantifiers are the number of times you want a pattern to be matched in order to match. 
Expressed as e{m,n}, where e is the expression or character we are matching, m is the minimum number of times you want it.

```python
# find the largest A streak a student recieved
re.findall("A{2,10}",grades)
```

```python
#  we can also find this using single values and just repeating the pattern
re.findall("A{1,1}A{1,1}", grades)
```

```python
# if we dont include a contifier the default is {1,1}
print(re.findall("AA", grades))



# if we have only one number in the braces, it's considered to be both m and n
print(re.findall("A{2}", grades))
```

```python
#using this we can find the decreasing trend of grades
re.findall("A{1,10}B{1,10}C{1,10}",grades)
```

##### Groups

Different patterns can be matched called groups using parentheses

```python
re.findall('([\w ]*)(\[edit\])',wiki)
```

Groups can be reffered using numbers as well

To return lists of match obects we can use the finditer function()

```python
for item in re.finditer('([\w ]*)(\[edit\])',wiki):
    print(item.groups())
```

We can get individual group using group(number), where group(0) is the whole match and each other number is the portion of the match we're intersted in

```python
for item in re.finditer('([\w ]*)(\[edit\])',wiki):
    print(item.group(1))
```

syntax for labeling or naming groups: ?P<name>

syntax to get a dictionary returned: .groupdict()

```python
#labelling or naming groups
#giving them a label and looking at the results in the dict can be done by using the syntax (?P<name>)
for item in re.finditer('(?P<title>[\w ]*)(?P<edit_link>\[edit\])',wiki):
    # we can get the dictionary returned for the item using .groupdict()
    print(item.groupdict()['title'])
```

```python
print(item.groupdict())
```

##### Look ahead and Look behind

```python
# the pattern being given to the regex engine is for the text either before or after the text we're trying to isolate
# isolate text before the [edit] and throwing the [edit]
# but if we want to use them to to match but not capture them we can put them in a group and use look ahead with the syntax ?=
for item in re.finditer('(?P<title>[\w ]+)(?=\[edit\])',wiki):
    print(item)


# what regex does is match two groups, first will be named and called title, we will have any amount of whitespace characters, the second will be the [edit] but we don't
# actually want this edit in our output match objects
```

### Group By

Pandas allows to iterate over every row in the dataframe called *groupby()*. The idea behind groupby() function is that it takes some dataframe, splits it into chunks based on some values, applies computation on those chunks, then combines the results back together into another dataframe. In pandas this is reffered to as the split-apply-combined pattern.

```python
df = pd.read_csv('C:/Users/chand/Desktop/datasets/census.csv')
df = df[df['SUMLEV']==50]
df.head()
```

```python
%%timeit -n 3
# get a list of unique states, then iterate over all states for each state, reduce the data frame and calculate the average
# run the function 3 times and %%timeit


for state in df['STNAME'].unique():
    # calculate average using numpy
    avg = np.average(df.where(df['STNAME']==state).dropna()['CENSUS2010POP'])
    print('Counties in state' + state +
          'have an average population of ' + str(avg))
```

```python
%%timeit -n 3
# for this methord we start by telling pandas we're interested in grouping by state name, this is called the split
for group, frame in df.groupby('STNAME'):
    # there are two values here, groupby() returns a tuple, where the first value is the value of the key we're trying to groupby, in this case that is the specific state name
    # and the second one is the projected dataframe that was found in the group
    
    # now we include out logic in the 'apply' step, which is to calculate an average of the census2010pop
    avg = np.average(frame['CENSUS2010POP'])
    # print the result
    print('Countries in state' + group +
          'have an average population of' + str(avg))
    
```

```python
# groupby can also use a function to segment the data

# in this fabricated example let's say that we wanted to work only with a third of the group batches at a time. Wwe could create some function to group and use that to 
# segment the data.

# we can create some new function called set_batch_number and if the first letter of the parameter is a capitel M we return a 0. If it's a capital Q we'll return a number 1
# and otherwise we'll return a 2. Then we'll pass this function to the data frame
df = df.set_index('STNAME')

def set_batch_number(item):
    if item[0]<'M':
        return 0
    if item[0]<'Q':
        return 1
    return 2

# loop through each batch 
for group, frame in df.groupby(set_batch_number):
    print('there are ' + str(len(frame)) + ' records in group ' + str(group) + 'for processing.')
```

##### Filtering

The groupby object has built in support for filtering groups as well. It can be used group by some feature, then make some transformation to the groups, then drop certain
groups as a part of your cleaning routines. The *filter()* function takes in a function which it applies to each grouped dataframe and returns either a True or a False, depending upon whether that group should be included in the results.

```python
# groups which have a mean rating above 9 included in our results
df.groupby('cancellation_policy').filter(lambda x:np.nanmean(x['review_scores_value'])>9.2)
```

note that our results are indexed, but that any of the results that were not a group with a mean score less tha or equal to 9.2 were not copied over.

##### Applying

The most common operation invoked on groupby(). This allows you to apply an arbitrary function to each group, and stitch back the results back for each apply() into a single dataframe where the index is preserved.

```python
df = pd.read_csv('C:/Users/chand/Desktop/datasets/listings.csv')

df = df[['cancellation_policy','review_scores_value']]
df.head()
```

```python
# in the previous examples we wanted to find the average review score of a listing and its deviation from the group mean. This was a two step process,
# first we used transform() on the groupby() object and then we had to broadcast to create a new dataframe. With apply we can wrap this logic in one place

def calc_mean_review_scores(group):
    # group is a dataframe of whatever we have grouped by e.g 'cancellation_policy', we can treat this as a df
    avg = np.nanmean(group['review_scores_value'])
    # now broadcast our formula and create a new column
    group['review_scores_mean']=np.abs(avg-group['review_scores_value'])
    return group

# apply to all groups
df.groupby('cancellation_policy').apply(calc_mean_review_scores).head()
```

Using apply can be a slower than using some of the specilized functions, especially *agg()*. But, if your dataframes are not huge, it's a very good general purpose approach

### Pivot Table

A pivot table is a way of summarizing data in a dataframe for particular purpose. It makes heavy use of the aggregation function. A pivot table itself a dataframe, where wthe rows represent one variable that you're interested in, the columns another, and the cell's some aggregate value. A pivot table also tends to include marginal values as well, which are sums for each column and row. This allows you to be able to see the relationship between two variables at just a glance.

```python
# Times higher education world university dataset, which is one of the most influencial university measures
df = pd.read_csv("C:/Users/chand/Desktop/datasets/university rankings/cwurData.csv")
df.head()
```

```python
def create_category(ranking):
    # since the ranking is an integer we can do an if-else
    if(ranking>=1) & (ranking <= 100):
        return 'First tier top uni'
    elif(ranking>=101) & (ranking <= 300):
        return 'Second tier top uni'
    elif(ranking>201) & (ranking <= 300):
        return 'Third tier top uni'
    else:
        return 'other top uni'
    
# now we apply this to a single column of data to create a new series
df['Rank_Level'] = df['world_rank'].apply(lambda x: create_category(x))
df.head()
```

```python
# a pivot table enables us to pivot one of these columns to a new column header and compare it against another column as row indices.
# compare rank = leval versus country of the universities

# to do this, we tell pandas we want the values to be Score, and the index to be the country and columns to be the rank levels. The we specify the aggregation function
# and here we use the numpy mean to get the average ratings

df.pivot_table(values='score', index='country', columns='Rank_Level', aggfunc=[np.mean, np.max]).head()
```

We notice NaN values, for example Argentina. The NaN values indicate that Argentina has only observations in the "Other Top Universities" category.

```python
df.pivot_table(values='score', index='country', columns='Rank_Level', aggfunc=[np.mean, np.max]).head(10)
```

```python
# overall average for the mean of the country for the mean and we want to see the max of the max, we can indicate that we want pandas to provide a marginal value
df.pivot_table(values='score', index='country', columns='Rank_Level', aggfunc=[np.mean, np.max],
               margins=True).head()
```

```python
# a pivot table is just a miltilevel dataframe, and we can access series or cells in a cell similar ways we do so for a regular dataframe.
new_df = df.pivot_table(values='score', index='country', columns='Rank_Level', aggfunc=[np.mean, np.max],
               margins=True)
# look at the index
print(new_df.index)
# look at the columns
print(new_df.columns)
```

The columns are hierarchial. The top level column indices indicate we have two categories: mean and max, and the lower level columns have four categories which are the four rank levels.

If we wanted to get the average scores for first top tier university level in each country, we would just need to make two dataframe projections, the first for the mean and the second for the top tier

```python
new_df['mean']['First tier top uni'].head()
```

Remember when you project a single column values out of a dataframe you get a series

```python
# if we want to find out the country that has the max average score in the Top tier uni level
# we can use the idxmax() function
new_df['mean']['First tier top uni'].idxmax()
```

### Basic statistical testing

Statistics is used in a lot if ways in data science. Hypothesis testing is a core data analysis activity behind experimentation. The goal behind hypothesis testing is to determine if, for instance, the two different conditions we have in an experiment have resulted in different impacts.

__*Scipy*__ includes numpy, pandas, matplotlib and a number of scientific functions as well

When we do hypothesis testing we have two statements of interest: the first is our actual explanation we have, which we call the alternative hypothesis, and the second is that the explanation we have is not sufficient, and we call this the null hypothesis, Our actual testing methord is to determine whether the null hypothesis is true or not. If we find that there is a difference between groups, then we can reject the null hypothesis and we can accept our alternative.

```python
df = pd.read_csv('C:/Users/chand/Desktop/datasets/class_grades.csv')
df.head()
```

```python
df = pd.read_csv("C:/Users/chand/Desktop/datasets/grades.csv")
df.head()
```

```python
# summary stats of the df
print(f'There are {df.shape[0]} and {df.shape[1]} columns')
```

```python
# let's segment this population into two segments. For those who finished the assignments by the end of Dec 2016, we'll call them early finishers, and for those who finish it 
# sometime after, call them late finishers.

early_finishers = df[pd.to_datetime(df['assignment1_submission']) < '2016']
early_finishers.head()
```

```python
#late_finishers = df[pd.to_datetime(df['assignment1_submission']) > '2016']

late_finishers = df[~df.isin(early_finishers.index)]
late_finishers.head()
```

```python
# compare the means of our two finishers

print(early_finishers['assignment1_grade'].mean())
print(late_finishers['assignment1_grade'].mean())
```

The above means are similar, but are they the same?? What do we mean by similar? This is where the student t-test comes in. It allows us to form an alternate hypothesis
('These are different') as well as the null hypothesis ('These are the same').

When we're doing hypothesis testing we must choose a significance level as a threshold for how much of a chance we're willing to accept . This significance level is typically called aplha.

For the example below we'll use a threshold of 0.05 for our alpha or 5%.

The SciPy library contains a number of different statistical tests and forms a basis for hypothesis testing in python and we're going to use the ttest_ind() function which does
an independent t-test (meaning the populations are not related to one another). The result of the ttest_index() are the t-statistic and a p-value. The later value which is the probablity, which is the most important to us, as it indicates the chance (between 0 and 1) of our null hypothesis being true.

```python
# import ttest_ind function
from scipy.stats import ttest_ind

ttest_ind(early_finishers['assignment1_grade'], late_finishers['assignment1_grade'])
```

So here we see that the probability is 0.469, and this is above our alpha of 0.05. This means that we cannot reject the null hypothesis. The null hypothesis was that the two populations are the same and we dont have enough certainty in our evidence(because it is greater than alpha) to come to a conclusion to the contrary. This does not mean that the two populations are the same.

```python
# check the other assignment grades
print(ttest_ind(early_finishers['assignment2_grade'], late_finishers['assignment2_grade']))
print(ttest_ind(early_finishers['assignment3_grade'], late_finishers['assignment3_grade']))
print(ttest_ind(early_finishers['assignment4_grade'], late_finishers['assignment4_grade']))
print(ttest_ind(early_finishers['assignment5_grade'], late_finishers['assignment5_grade']))
print(ttest_ind(early_finishers['assignment6_grade'], late_finishers['assignment6_grade']))
```

It looks like in this data we do not have enough evidence to suggest the populations differ with respect to grade.

*One issue with p-value is that you run more tests you are likely to get a value which is startistically significant just by chance.*
